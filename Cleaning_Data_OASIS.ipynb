**CLEANING DATA**

Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**DATA LOADING**

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

```python
print("Shape of dataset:", df.shape)
```

```python
df.info()
```

```python
df.describe(include='all')
```

**DATA CLEANING**

Handling Missing Values

```python
df.isnull().sum()
```

```python
df["reviews_per_month"] = df["reviews_per_month"].fillna(0)
```

```python
df.dropna(subset=["name", "host_name"], inplace=True)
```

**REMOVING DUPLICATES**

```python
print("Duplicates before cleaning:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("Duplicates after cleaning:", df.duplicated().sum())
```

**STANDARIZING FORMATS**

```python
df["last_review"] = pd.to_datetime(df["last_review"], errors="coerce")
```

```python
df["neighbourhood_group"] = df["neighbourhood_group"].str.lower()
df["neighbourhood"] = df["neighbourhood"].str.lower()
df["room_type"] = df["room_type"].str.lower()
```

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce").astype(int)
```

```python
df["minimum_nights"] = pd.to_numeric(df["minimum_nights"], errors="coerce").astype(int)
```

**DETECTING OUTLIERS**

Price vs. Minimum Nights

```python
plt.figure(figsize=(8,6))
plt.scatter(df["minimum_nights"], df["price"], alpha=0.3)
plt.xlabel("Minimum Nights")
plt.ylabel("Price ($)")
plt.title("Scatter Plot: Price vs Minimum Nights (with Outliers)")
plt.show()
```

Price vs. Number of Reviews

```python
plt.figure(figsize=(8,6))
plt.scatter(df["number_of_reviews"], df["price"], alpha=0.3, c="orange")
plt.xlabel("Number of Reviews")
plt.ylabel("Price ($)")
plt.title("Scatter Plot: Price vs Number of Reviews (with Outliers)")
plt.show()
```

Zoomed in Price vs. Minimum Nights

```python
plt.figure(figsize=(8,6))
subset = df[df["minimum_nights"] <= 100]   # zoom in
plt.scatter(subset["minimum_nights"], subset["price"], alpha=0.3, c="green")
plt.xlabel("Minimum Nights (<=100)")
plt.ylabel("Price ($)")
plt.title("Scatter Plot: Price vs Minimum Nights (Zoomed In)")
plt.show()
```

HANDLING OUTLIERS

```python
df_cleaned = df.copy()
df_cleaned = df_cleaned[(df_cleaned["price"] > 0) & (df_cleaned["price"] <= 1000)]
df_cleaned = df_cleaned[df_cleaned["minimum_nights"] <= 365]
print("Rows before:", df.shape[0])
print("Rows after fixing outliers:", df_cleaned.shape[0])
print("Price range after cleaning:", df_cleaned["price"].min(), "-", df_cleaned["price"].max())
print("Minimum nights range after cleaning:", df_cleaned["minimum_nights"].min(), "-", df_cleaned["minimum_nights"].max())
```

```python
plt.figure(figsize=(10,6))
scatter = plt.scatter(
    df_cleaned["minimum_nights"], 
    df_cleaned["price"], 
    c=df_cleaned["price"],   
    cmap="viridis",         
    alpha=0.6, 
    edgecolor="k", 
    s=50                     
)
plt.title("Scatter Plot: Price vs Minimum Nights (After Fixing Outliers)", fontsize=16, fontweight="bold")
plt.xlabel("Minimum Nights", fontsize=14)
plt.ylabel("Price ($)", fontsize=14)
cbar = plt.colorbar(scatter)
cbar.set_label("Price ($)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
df_cleaned = df.copy()
df_cleaned = df_cleaned[(df_cleaned["price"] > 0) & (df_cleaned["price"] <= 1000)]
df_cleaned = df_cleaned[df_cleaned["number_of_reviews"] <= 500]
print("Rows before:", df.shape[0])
print("Rows after:", df_cleaned.shape[0])
print("Price range:", df_cleaned["price"].min(), "-", df_cleaned["price"].max())
print("Number of reviews range:", df_cleaned["number_of_reviews"].min(), "-", df_cleaned["number_of_reviews"].max())
```

```python
room_type_colors = {
    "entire home/apt": "red",
    "private room": "blue",
    "shared room": "green",
    "hotel room": "purple"
}
plt.figure(figsize=(10,7))
for room_type, color in room_type_colors.items():
    subset = df_cleaned[df_cleaned["room_type"] == room_type]
    plt.scatter(
        subset["number_of_reviews"],
        subset["price"],
        alpha=0.4,
        c=color,
        label=room_type,
        edgecolors="w",
        s=60
    )
plt.xlabel("Number of Reviews (<= 500)", fontsize=12, fontweight="bold")
plt.ylabel("Price ($)", fontsize=12, fontweight="bold")
plt.title("Price vs Number of Reviews (After Fixing Outliers)", fontsize=14, fontweight="bold")
plt.legend(title="Room Type")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
df_fixed = df.copy()
df_fixed = df_fixed[(df_fixed["price"] > 0) & (df_fixed["price"] <= 1000)]
df_fixed = df_fixed[df_fixed["minimum_nights"] <= 100]
print("Rows before fixing:", df.shape[0])
print("Rows after fixing:", df_fixed.shape[0])
```

```python
plt.figure(figsize=(10,6))
scatter = plt.scatter(
    df_fixed["minimum_nights"], 
    df_fixed["price"], 
    c=df_fixed["price"], 
    cmap="plasma",       
    alpha=0.6, 
    edgecolor="k", 
    s=50
)
plt.title("Price vs Minimum Nights (After Fixing Outliers, Zoomed 100)", fontsize=16, fontweight="bold")
plt.xlabel("Minimum Nights ( 100)", fontsize=14)
plt.ylabel("Price ($)", fontsize=14)
cbar = plt.colorbar(scatter)
cbar.set_label("Price ($)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python

summary = pd.DataFrame({
    "Metric": ["Row Count", "Missing Values", "Duplicate Rows", "Max Price", "Max Minimum Nights"],
    "Before Cleaning": [
        df.shape[0],
        df.isnull().sum().sum(),
        df.duplicated().sum(),
        df["price"].max(),
        df["minimum_nights"].max()
    ],
    "After Cleaning": [
        df_fixed.shape[0],
        df_fixed.isnull().sum().sum(),
        df_fixed.duplicated().sum(),
        df_fixed["price"].max(),
        df_fixed["minimum_nights"].max()
    ]
})

print("Data Cleaning Summary")
display(summary)
```
